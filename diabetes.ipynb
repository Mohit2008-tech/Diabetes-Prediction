```python
q = df['Pregnancies'].quantile(.98)    #Remove the 2% data of Pregnancies
df_new = df[df['Pregnancies']<q]
```

```python
q = df['BMI'].quantile(.99)            #Remove the 1% data of BMI
df_new = df_new[df_new['BMI']<q]
```

```python
q = df['SkinThickness'].quantile(.99)            #Remove the 1% data of SkinThickness
df_new = df_new[df_new['SkinThickness']<q]
```

```python
q = df['DiabetesPedigreeFunction'].quantile(.99)            #Remove the 1% data of DiabetesPedigreeFunction
df_new = df_new[df_new['DiabetesPedigreeFunction']<q]
```

```python
q = df['Age'].quantile(.99)            #Remove the 1% data of Age
df_new = df_new[df_new['Age']<q]
```

```python
df_new
```

```python
#define x and y
```

```python
x = df_new.drop(columns=['Outcome'])
```

```python
x
```

```python
y = df_new['Outcome']
```

```python
y
```

```python
#Standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
x_scaled.shape[1]
```

```python
x_df = pd.DataFrame(x_scaled,columns = x.columns)
x_df
```

```python
#checking Multicollinearity using VIF score
from statsmodels.stats.outliers_influence import variance_inflation_factor
def vif_score(x):
    scaler = StandardScaler()
    arr = scaler.fit_transform(x)
    return pd.DataFrame([[x.columns[i] , variance_inflation_factor(arr,i)] for i in range(arr.shape[1])] , columns  = ['Features','VIF_SCORE'])
```

```python
vif_score(x)
```

```python
#Training and Testing
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x_df , y , test_size = .20 , random_state = 144)
```

```python
#Logistic Regression 
from sklearn.linear_model import LogisticRegression
logr = LogisticRegression()
logr.fit(x_train , y_train)
```

```python
logr.predict(x_test)
```
